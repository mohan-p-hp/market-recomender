```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv('data/market_prices.csv')
df['date'] = pd.to_datetime(df['date'])

# Basic info about your data
print("Dataset shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst few rows:")
print(df.head())

# Plot price trends for Tomato
plt.figure(figsize=(12, 6))
for market in df['market_name'].unique():
    market_data = df[df['market_name'] == market]
    tomato_data = market_data[market_data['commodity'] == 'Tomato']
    plt.plot(tomato_data['date'], tomato_data['modal_price'], label=market)

plt.title('Tomato Prices Across Different Markets')
plt.xlabel('Date')
plt.ylabel('Price (₹)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
